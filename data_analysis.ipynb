```python
import pandas as pd
import numpy as np
import ipynb
from ipynb.fs.full.sfacleaner import yearconv
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pklpath = r'/home/u1033387/FinalProject/pickles/'
hdm = pd.read_pickle(pklpath+'hdmaster.pkl')
```

```python
# for each year, run the cleaning functions housed in the 'data_parsing_and_cleaning.ipynb' notebook, and name all the columns. create a new df
years = [x for x in range(2020, 1999, -1)]
thisdf = pd.DataFrame(columns=['YEAR','SCUGRAD','UPGRNTN','UPGRNTT','UPGRNTA','UFLOANN','UFLOANT','UFLOANA','FTIS','FTOS','PTIS (9CH)','PTOS (9CH)'])
for year in years:
    thissfa = pd.read_pickle(pklpath+'sfa{}.pkl'.format(yearconv(year)))
    thisic = pd.read_pickle(pklpath+'ic{}_ay.pkl'.format(year))
    thiscombo = pd.merge(thissfa, thisic, on=['UNITID','YEAR'])
    thisavg = thiscombo.drop('UNITID', axis=1)
    thisavg = thisavg.mean()
    thisavgdf = thisavg.to_frame()
    thisdf = thisdf.append(thisavgdf.transpose())
```

```python
# save to pickle
thisdf.to_pickle(pklpath+'allavg.pkl')
```

```python
# read pickle back in
updf= pd.read_pickle(pklpath+'allavg.pkl')
updf = updf.reset_index()
updf = updf.drop('index', axis=1)
```

```python
# rename all columns for presentation
def avgcol(file):
    avgcoldict = {'YEAR':'YEAR',
'SCUGRAD':'# UGRAD STUDENTS',
'UPGRNTN':'# PGRANT RECIPIENTS',
'UPGRNTT':'TOTAL PGRANT $',
'UPGRNTA':'AVG. PGRANT $/RECIPIENT',
'UFLOANN':'# LOAN RECIPIENTS',
'UFLOANT':'TOTAL LOAN $',
'UFLOANA':'AVG. LOAN $/RECIPIENT' ,
'FTIS':'FULL-TIME IN STATE COST',
'FTOS':'FULL-TIME OUT OF STATE COST',
'PTIS (9CH)':'PART-TIME IN STATE COST (9CH)',
'PTOS (9CH)':'PART-TIME OUT OF STATE COST (9CH)'}
    for column in avgcoldict:
        for x in range(len(avgcoldict)):
            if column == file.columns.values[x]:
                file.columns.values[x] = avgcoldict[column]
    return file
```

```python
# create new pickle containing average amounts for each year
updf = avgcol(updf)
updf = updf.sort_values('YEAR')
updf.to_pickle(pklpath+'allavg.pkl')
```
